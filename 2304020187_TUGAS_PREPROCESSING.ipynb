# CEK DAN PENANGANAN OUTLIER

**1. Boxplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kolom numerik yang ingin dicek outliernya
num_cols = ['Attendance']

# Visualisasi Boxplot
plt.figure(figsize=(10, 6))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

Berdasarkan hasil boxplot terlihat beberapa titik di luar batas normal yang menunjukkan adanya outlier yaitu data dengan nilai yang jauh berbeda dari sebagian besar data lainnya.

**2. IQR**

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Kuartil 1
    Q3 = data[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

Berdasarkan hasil metode IQR, diperoleh bahwa kolom Attendance memiliki 37 data yang termasuk outlier yaitu nilai yang berada di luar batas normal

**- Winsorizing (Mengganti Outlier dengan Batas IQR)**

```python
def winsorize_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# Terapkan Winsorizing
for col in num_cols:
    winsorize_iqr(data, col)
```

```python
#cek boxplot lagi

num_cols = ['Attendance']

# Visualisasi Boxplot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

Berdasarkan penanganan outlier menggunakan metode winsorizing, nilai yang berada di luar batas normal diganti dengan nilai batas bawah dan batas atas. Hasil visualisasi boxplot menunjukkan bahwa outlier sudah tidak terlihat, sehingga data menjadi lebih stabil untuk analisis
